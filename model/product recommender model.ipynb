```python
import re
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier as KNN 
```

```python
df = pd.read_csv('cleaned_sephora_dataset.csv')
```

```python
df.head()
```

```python
bins = [0,25,50,250]
labels = ['<$25' , '$25-$50', '>$50']
df['budget'] = pd.cut(df['price'], bins, labels =labels)
```

```python
df['budget']
```

```python
df_2 = df['budget'].str.join('').str.get_dummies()
df= df.join(df_2).drop('budget', axis = 1)
```

```python
df = df.drop('and', 1)
```

```python
df.Label.unique()
```

```python
df.Label[df['Label']=='Face Wash'] = str('0')
df.Label[df['Label']=='Moisturizer'] = str('1')
df.Label[df['Label']=='Shaving Kits'] = str('2')
df.Label[df['Label']=='Fragrance'] = str('3')
df.Label[df['Label']=='Eye Cream'] = str('4')
df.Label[df['Label']=='Deodorant'] = str('5')
df.Label[df['Label']=='Hair products'] = str('6')
```

```python
df.head()
```

```python
df.columns[11:22].tolist()
```

```python
df1 = df[['Label','brand_name','product_name','price','score','Combination', \
         'Dry','Normal','Oily','Sensitive']]
df1.head()
```

```python
df1['name'] = df1[['brand_name','product_name']].apply(lambda x: ' '.join(x), axis =1)
df1['name'].head()
```

```python
X = df1[['Label','price','score','Combination', \
         'Dry','Normal','Oily','Sensitive']]
```

```python
y = df1['name']
```

```python
# Split dataset into train and test 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, 
                        random_state = 2018)
```

```python
knn = KNN(n_neighbors = 1)
```

```python
# train model 
knn.fit(X_train, y_train)
```

```python
import pickle 
  
# Save the trained model as a pickle string. 
saved_model = pickle.dumps(knn) 
  
# Load the pickled model 
knn_from_pickle = pickle.loads(saved_model) 
  
# Use the loaded pickled model to make predictions 
knn_from_pickle.predict(X_test)
```
